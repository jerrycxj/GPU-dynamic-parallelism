```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3,3, figsize=(15,15))
fig.figsize = ((16,15))
for i in range(3):
    for j in range(3):
        axs[i, j].plot(sizes, final[:,i,2,j,:])
        axs[i, j].title.set_text("Max dwell: "+str(maxdwell[i]) + "\nMin Size: " + str(minSize[j]))
```
