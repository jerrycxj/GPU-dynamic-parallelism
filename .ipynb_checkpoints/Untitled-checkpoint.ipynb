```python
sizes = [2**i*1024 for i in np.arange(5)]
for i in range(11):
    sizes.append((18+2*i)*1024 )
sizes
```

```python
final = np.zeros((16,3,3,3,3))
cont = 0
for i in sizes:
    for j in range(3):
        dummy = !make -B -C /home/newt/Documents/mandelbrot-dyn/mandelbrot/ H=$(({i})) W=$(({i})) MAX_DWELL=$((2**(7+{j})))
        res1 = 0
        for u in range(10):
            r = !exec "/home/newt/Documents/mandelbrot-dyn/mandelbrot/mandelbrot"
            res1 += float(r[0])
        res1 /= 10
        print("Ended Mandelbrot normal")
        
        for k in range(3):
            for l in range(3):
                dummy = !make -B -C '/home/newt/Documents/mandelbrot-dyn/mandelbrot-1kernel/' H={i} W={i} MAX_DWELL=$((2**(7+{j}))) MAX_DEPTH=$(({k}+3)) MIN_SIZE=$((2**(4+{l}))) 
                dummy = !make -B -C '/home/newt/Documents/mandelbrot-dyn/mandelbrot-dyn/'     H={i} W={i} MAX_DWELL=$((2**(7+{j}))) MAX_DEPTH=$(({k}+3)) MIN_SIZE=$((2**(4+{l}))) 
                res2 = 0
                res3 = 0
                for u in range(10):
                    r = !exec "/home/newt/Documents/mandelbrot-dyn/mandelbrot-1kernel/mandelbrot-1ker"
                    res2 += float(r[0])
                print("Ended Mandelbrot 1kernel")
                for u in range(10):
                    r = !exec "/home/newt/Documents/mandelbrot-dyn/mandelbrot-dyn/mandelbrot-dyn"
                    res3 += float(r[0])
                print("Ended Mandelbrot dynamic")
                res2 /= 10
                res3 /= 10

                final[cont,j,k,l, 0] = res1
                final[cont,j,k,l, 1] = res2
                final[cont,j,k,l, 2] = res3
                print("Actual", cont, j, k, l)
    cont+=1
```
